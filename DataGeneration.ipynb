```python
from collections import defaultdict
```

```python
def read_qa(files):
    res = defaultdict(list)
    for file_nm in files:
        f = open(file_nm, 'r')
        topics = []
        questions = []
        answers = []
        for i, line in enumerate(f):
            line = line.rstrip()
            if line:
                if line[0] == 'T':
                    topics.append(line[3:])
                if line[0] == 'Q':
                    questions.append(line[3:])
                if line[0] == 'A':
                    answers.append(line[3:])
        f.close()
        for t, q, a in zip(topics, questions, answers):
            res[t].append((q, a))
    
    return res
```

```python
files = ['data/premise_question_answer4.txt', 'data/premise_question_answer5.txt',
        'data/premise_question_answer6.txt']
res = read_qa(files)
```

```python
def generate_tasks_assignments(qa_data, group_ids):
    assignment_groups = []
    assignment_description = []
    assignment_dates = []
    tasks = []
    assignments = list(qa_data.keys())
    print('OK')
    for group_id in group_ids:
        num_assignments = np.random.randint(5) + 3
        descriptions = list(np.random.choice(assignments, num_assignments, replace=False))
        assignment_description += descriptions
        for assign in range(num_assignments):
            assignment_groups.append(group_id)
            day = np.random.randint(31) + 1
            date = '2020-01-' + str(day).rjust(2, '0') + ' 23:00:00'
            assignment_dates.append(date)
        if (group_id + 1) % 500 == 0:
            print(group_id)
            
    assignment_df = pd.DataFrame({'assignment_id': np.arange(1, len(assignment_groups)+1),
                                  'group_id' : assignment_groups,
                                  'description': assignment_description,
                                  'due_date': assignment_dates})
    return assignment_df
```

```python
def generate_tasks(assignment_df, qa_data):
    task_assignments = []
    task_groups = []
    task_descriptions = []
    task_answers = []
    for i, row in assignment_df.iterrows():
        qas = qa_data[row['description']]
        num_tasks = np.random.randint(1, min(len(qas) + 1, 5))
        q_a_ind = list(np.random.choice(len(qas), num_tasks, replace=False))
        q_a = [text for j, text in enumerate(qas) if (j in q_a_ind)]
        questions = [j[0] for j in q_a]
        answers = [j[1] for j in q_a]
        task_groups += ([row['group_id']] * num_tasks)
        task_assignments += ([row['assignment_id']] * num_tasks)
        task_descriptions += questions
        task_answers += answers
        if (i + 1) % 1000 == 0:
            print(i)

    tasks_df = pd.DataFrame({'task_id': np.arange(len(task_assignments))+1,
                             'description': task_descriptions,
                             'assignment_id': task_assignments,
                             'group_id' : task_groups,
                             'answers': task_answers})
    return tasks_df    
```

```python
def generate_submissions(tasks_df, students_df, all_answers):
    task_ids = []
    student_ids = []
    dates = []
    contents = []
    group_ids = []
    for i, row in tasks_df.iterrows():
        group = row['group_id']
        students_group = list(students_df.loc[students_df['group_id'] == group]['student_id'].values)
        # 30-80 % solved task
        num_students = np.random.randint(int(0.3 * len(students_group)),
                                         int(0.8 * len(students_group)))
        group_ids += [group] * num_students
        task_ids += [row['task_id']] * num_students
        dates += ['2019-12-' + str(np.random.randint(31)+1).rjust(2, '0')
                  + ' 22:00:00' for j in range(num_students)]
        student_ids += list(np.random.choice(students_group, num_students, replace=False))
        for j in range(num_students):
            if np.random.randint(10) < 7:
                answer = row['answers']
            else:
                answer = np.random.choice(all_answers, 1)[0]
            contents.append(answer)
        if (i + 1) % 1000 == 0:
            print(i)
        
    submissions_df = pd.DataFrame({'submission_id': np.arange(len(task_ids))+1,
                             'task_id': task_ids,
                             'student_id': student_ids,
                             'date': dates,
                             'content': contents,
                             'group_id' : group_ids})
    return submissions_df
```

```python
def generate_supervisors(submissions_df, prepods_df):
    teacher_ids = []
    submission_ids = []
    for i, row in submissions_df.iterrows():
        group = row['group_id']
        teachers_group = list(prepods_df.loc[prepods_df['group_id'] == group]['teacher_id'].values)
        # 0-3 supervisors
        num_supervisors = min(np.random.randint(4), len(teachers_group))
        if num_supervisors > 0:
            submission_ids += [row['submission_id']] * num_supervisors
            teacher_ids += list(np.random.choice(teachers_group, num_supervisors, replace=False))
        if (i + 1) % 10000 == 0:
            print(i) 
    
    supervisors_df = pd.DataFrame({'supervisor_id': np.arange(len(teacher_ids))+1,
                                   'teacher_id': teacher_ids,
                                   'submission_id' : submission_ids})
    return supervisors_df
```

```python
def generate_grades(supervisors_df, submissions_df):
    grades = []
    dates = []
    supervisor_ids = []
    submission_ids = []
    for i, row in supervisors_df.iterrows():
        num_grades = np.random.randint(3)
        min_date = submissions_df.loc[submissions_df['submission_id'] == row['submission_id']]['date'].values[0]
        min_day = int(min_date[8:10])
        if num_grades > 0:
            supervisor_ids += [row['supervisor_id']] * num_grades
            submission_ids += [row['submission_id']] * num_grades
            for j in range(num_grades):
                grade = np.random.randint(11)
                grade = grade + np.random.randint(10 - grade + 1)
                grades.append(grade)
                day = np.random.randint(min_day, 32)
                date = '2019-12-' + str(day).rjust(2, '0') + ' 23:00:00'
                dates.append(date)
        if (i + 1) % 10000 == 0:
            print(i) 
            
    grades_df = pd.DataFrame({'grade_id': np.arange(len(grades))+1,
                              'supervisor_id': supervisor_ids,
                              'grade': grades,
                              'date' : dates})
    return grades_df
            
```
